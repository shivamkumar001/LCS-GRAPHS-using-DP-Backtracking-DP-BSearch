```
#worst case
import matplotlib.pyplot as plt 

# line 1 points 
x1 = [0,25,50,75,100,110,120,130,140,145] 
y1 = [0,0,2,4,75,196,748,2022,4244,5702] 
# plotting the line 1 points 
plt.plot(x1, y1, color='green',label = "Backtracking with Recursion") 

# line 2 points 
x2 = [0,50,500,1000,5000,7000,10000,15000,30000] 
y2 = [0,0,2,5,120,276,503,1093,4390] 
# plotting the line 2 points 
plt.plot(x2, y2,color='magenta', label = "DP Memoization") 

# line 3 points 
x3 = [0,50,500,1000,5000,7000,10000,15000,30000] 
y3 = [0,1,2,4,93,179,368,820,3401] 
# plotting the line 1 points 
plt.plot(x3, y3,color='blue', label = "DP with Tabulation") 

# line 4 points 
x4 = [0,5000,10000,30000] 
y4 = [0,2,4,8] 
# plotting the line 1 points 
plt.plot(x4, y4,color='cyan', label = "Divide-and-Conquer with Recursion") 

# naming the x axis 
plt.xlabel('Input Size') 
# naming the y axis 
plt.ylabel('Running Time (ms)') 
# giving a title to my graph 
plt.title('WORST CASE') 

# show a legend on the plot 
plt.legend() 

# function to show the plot 
plt.show() 

```
